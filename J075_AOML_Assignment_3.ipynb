## Cat in the Dat

```python
df = pd.read_csv('C:/Users/Yatha/OneDrive/Documents/Sem VI/AOML/cat-in-the-dat-ii/train.csv')
df.head()
```

```python
df.columns
```

```python
# Ordinal Encoding
for col in df.columns:
    if col.startswith('ord'):
        print(col, df[col].unique())
        print()
```

```python
# ord_0 is already ordinal

ord_1_label_map = {
    "Novice": 0,
    "Contributor": 1,
    "Expert": 2,
    "Master": 3,
    "Grandmaster": 4
}

df["ord_1_ordered_mapping"] = df["ord_1"].map(ord_1_label_map)

ord_2_label_map = {
    "Freezing": 0,
    "Cold": 1,
    "Warm": 2,
    "Hot": 3,
    "Boiling Hot": 4,
    "Lava Hot": 5
}

df["ord_2_ordered_mapping"] = df["ord_2"].map(ord_2_label_map)

def alphabetical_ordering(col):
    col = list(col.unique())
    col.remove(np.nan)
    col.sort()
    return {k: i for i, k in enumerate(col)}

# Sorting categories alphabetically
ord_3_label_map = alphabetical_ordering(df["ord_3"])
df["ord_3_ordered_mapping"] = df["ord_3"].map(ord_3_label_map)

ord_4_label_map = alphabetical_ordering(df["ord_4"])
df["ord_4_ordered_mapping"] = df["ord_4"].map(ord_4_label_map)

ord_5_label_map = alphabetical_ordering(df["ord_5"])
df["ord_5_ordered_mapping"] = df["ord_5"].map(ord_5_label_map)
```

```python
df.head()
```

```python
# Filling missing values
for col in df.columns:
    if col.startswith('ord') or col.startswith('nom'):
        df[col] = df[col].fillna('UNKNOWN')
```

```python
ord_1_freq = df["ord_5"].value_counts()
df["ord_5_frequency"] = df["ord_5"].map(ord_1_freq)
df["ord_5_is_rare"] = df["ord_5_frequency"] < 1000
```

```python
df['day'] = df['day'].fillna(df['day'].mode()[0])
df['month'] = df['month'].fillna(df['month'].mode()[0])
```

```python
df["month"] = df["month"].fillna(df["month"].mode()[0]).astype(int)
df["nom_3"] = df["nom_3"].fillna(df["nom_3"].mode()[0])

df["month_and_nom_0"] = df["month"].astype(str) + "_" + df["nom_3"].astype(str)
```

```python
import itertools

cols_for_concatenation = ["nom_1", "nom_2", "nom_3", "nom_4"]
col_combinations = list(itertools.combinations(cols_for_concatenation, 2))

col_combinations
```

```python
deg2_cols = []

for col_a, col_b in col_combinations:
    new_col = "{}_{}".format(col_a, col_b)
    df[new_col] = df[col_a].astype(str) + "_" + df[col_b].astype(str)
    deg2_cols.append(new_col)
```

```python
# Feature Engineering for cyclical - sin and cos
cycle_cols = ['day', 'month']
for col in cycle_cols:
    df[col + '_sin'] = np.sin(2 * np.pi * df[col]/df[col].max())
    df[col + '_cos'] = np.cos(2 * np.pi * df[col]/df[col].max())
    df.drop(col, axis=1, inplace=True)
```

```python
for col in df.columns:
    if col.startswith('bin'):
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
for col in df.columns:
    if col!='target':
        df[col] = df[col].astype('category')
```
